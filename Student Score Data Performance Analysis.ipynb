# Student Score Data Performance Analysis

Problem: Whose homework performed the best? Which homework can students performed best on?

Approach: Using pandas to clean datasets and format, rank averge score by person and by homework.

Result: HW10 is the one students get highest score among all those homeworks. And yang performed the best on homework. 

```python
import pandas as pd
pd.set_option("display.precision", 2)
df_HW = pd.read_csv("HW_data.csv")
df_HW
```

## split the name into English and Chinese part, delete the original column

```python
df_HW['English_Name']=df_HW['Name'].str.rsplit(n=1,expand = True)[0]
df_HW['Chinese_Name']=df_HW['Name'].str.rsplit(n=1,expand = True)[1]
df_HW.drop(columns=['Name'],inplace=True)
df_HW
```

## set the Chinese name as index

```python
df_HW.set_index('Chinese_Name',inplace=True)
df_HW
```

## Check the score for '王军琴' at 'HW10'

```python
df_HW.loc['王军琴','HW10']
```

## Create a subset of data with names starting from 李 and for HW5 to HW10

```python
df_HW[df_HW.index.str.startswith('李')].loc[:,'HW5':'HW10']
```

## Missing value of each homework exploration
Count how many HWs are missing for each HW. Find HW13 has the most missing.

```python
NA_HW=df_HW.isnull().sum()
NA_HW
```

```python
NA_HW.idxmax()
```

## Missing value of each student exploration

```python
NA_person = df_HW.isnull().sum(axis = 1)
NA_person
```

```python
NA_person.idxmax()
```

## Average score of each homework
Calculate the average score for each HW and save it into a new row

```python
new_row=pd.Series(data=df_HW.mean())
new_data=pd.DataFrame([new_row],index=['Average_HW'])
df_HW=df_HW.append(new_data,sort=True)
df_HW
```

## Rank the average score for each HW from highest to lowest

```python
df_HW.loc['Average_HW'].sort_values(ascending=False)
```

## Average score for each student 

```python
df_HW['Average_person']=df_HW.mean(axis=1)
df_HW
```

## Rank the average score for each person from highest to lowest

```python
df_HW['Average_person'].sort_values(ascending=False)
```
